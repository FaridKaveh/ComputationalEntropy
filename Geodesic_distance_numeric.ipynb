```python
def phi(x: float, n: int) -> float:
    if x == 1:
        return (n - 1) / 2
    else:
        nom = x - n * (x**n) + (n - 1) * (x ** (n + 1))
        denom = (1 - x) * (1 - x**n)

        return nom / denom

def varphi(x, n, delta):
    return phi(1 / delta * mp.exp(-x), n)
```
